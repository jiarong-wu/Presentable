```python
list = [[200,0.13,0.35],[200,0.17,0.75],[200,0.214,1.40],[200,0.248,2.15],
        [300,0.13,0.30],[300,0.17,0.80],[300,0.214,1.70],[300,0.248,2.60],
        [400,0.13,0.20],[400,0.17,0.75],[400,0.214,1.75],[400,0.248,2.90],
                        [500,0.17,0.50],[500,0.214,1.74],[500,0.248,2.89],
                        [600,0.17,0.10],[600,0.214,1.40],[600,0.248,2.75],
                                        [700,0.214,1.00],[700,0.248,2.40],
                                        [800,0.214,0.40],[800,0.248,1.78],
                                                         [900,0.248,1.12]]
realwave = RealWave(g = 9.8, sigma = 0.074, rho = 1000, rho_air = 1.225)
for item in list:
    realwave.k2omega(item[0])
    # wave frequency
    item.append(realwave.omega)
    # non-dimensional growth rate
    item.append(item[2]/realwave.omega)
    # wave speed
    item.append(realwave.c)
    # non-dimensional u*
    item.append(item[1]/realwave.c)
    # Bo number
    item.append(realwave.Bo)
    
print(list)
```

```python
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

fig = plt.figure()
ax1 = fig.gca()
Bo = []
nondim_growth = []
nondim_ustar = []
size = 150
for i in [3,7,11,14,17,19,21,22]:
    item = list[i]
    nondim_growth.append(item[4])
    Bo.append(item[7])
    nondim_ustar.append(item[6])
ax1.scatter(Bo, nondim_growth, s=size, marker = '*', facecolors='none', edgecolors = cycle[0])
for i, ustar in enumerate(nondim_ustar):
    ax1.annotate(r'\textbf{%1.2f}' %ustar, (Bo[i], nondim_growth[i]))
Bo = []
nondim_growth = []
nondim_ustar = []
for i in [2,6,10,13,16,18,20]:
    item = list[i]
    nondim_growth.append(item[4])
    Bo.append(item[7])
    nondim_ustar.append(item[6])
ax1.scatter(Bo, nondim_growth, s=size, marker = '*', facecolors='none', edgecolors = cycle[1])
for i, ustar in enumerate(nondim_ustar):
    ax1.annotate(r'\textbf{%1.2f}' %ustar, (Bo[i], nondim_growth[i]))
Bo = []
nondim_growth = []
nondim_ustar = []
for i in [1,5,9,12,15]:
    item = list[i]
    nondim_growth.append(item[4])
    Bo.append(item[7])
    nondim_ustar.append(item[6])
ax1.scatter(Bo, nondim_growth, s=size, marker = '*', facecolors='none', edgecolors = cycle[2])
for i, ustar in enumerate(nondim_ustar):
    ax1.annotate(r'\textbf{%1.2f}' %ustar, (Bo[i], nondim_growth[i]))
Bo = []
nondim_growth = []
nondim_ustar = []
for i in [0,4,8]:
    item = list[i]
    nondim_growth.append(item[4])
    Bo.append(item[7])
    nondim_ustar.append(item[6])
ax1.scatter(Bo, nondim_growth, s=size, marker = '*', facecolors='none', edgecolors = cycle[3])
for i, ustar in enumerate(nondim_ustar):
    ax1.annotate(r'\textbf{%1.2f}' %ustar, (Bo[i], nondim_growth[i]))

ax1.set_ylabel(r'$\overline{\beta} = \beta/\omega$')
plt.locator_params(axis='y', nbins=6)
ax1.set_xlabel(r'Bo = $\Delta\rho g /\sigma k^2$')
ax1.set_title(r'\textbf{Non-dim growth rate vs. Bo}')
ax1.set_xscale('log')
ax1.set_xticks([0.27,0.53,1.47,3.31])
import matplotlib
ax1.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax1.set_xlim([0.15, 4])
ax1.set_ylim([0,0.04])
fig.savefig('/home/jiarong/research/presentation/APSDFD/comparison1.pdf', bbox_inches='tight')
```
